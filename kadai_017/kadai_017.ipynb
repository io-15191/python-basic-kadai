```python
class Human:
  #コントラクタ
  def __init__(self, name, age):
    #属性を定義する
        self.name = name
        self.age = age
  #条件分岐のメソッド
  def check_adult (self):
              if self.age >= 20:
                print(f"{self.name} は大人です")
              else:
                print (f"{self.name}は大人ではありません。")
#インスタンス化
suzuki = Human("suzuki",36)
sato = Human("sato", 15)
tanaka = Human("tanaka",23)
ooishi = Human("ooishi",45)

human_list = [suzuki,sato,tanaka,ooishi]

for human in human_list:
  human.check_adult()


```
